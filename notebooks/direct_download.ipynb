```python
zip_2023= "https://content.explore-education-statistics.service.gov.uk/api/releases/99b7ca69-f50a-42b2-aeda-b99475a54c31/files/"
r_2023 = requests.get(zip_2023)
```

```python
zip_2023 = zipfile.ZipFile(io.BytesIO(r_2023.content))
```

```python
for x in zip_2023.namelist():
    print(x)
```

```python
for file_name  in zip_2023.namelist():
    print(file_name)
    f = zip_2023.extract(file_name)
    df = pd.DataFrame()
    try:
        df = pd.read_csv(f, encoding='utf-8',delimiter=',', encoding_errors='replace')
    except UnicodeDecodeError:
        df = pd.read_csv(f, encoding='latin1',delimiter=',', encoding_errors='replace')  # Try a different encoding
    except pd.errors.ParserError:
        print(f"Error parsing {file_name}")
    if not df.empty: print(df.columns)
    continue

    
```
